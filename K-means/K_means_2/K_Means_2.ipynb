# Task 1: Scatter Plot Showing Initial Points

```python
plt.scatter(data['ApplicantIncome'],data['LoanAmount'],c = 'blue')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount (in thousands)')
plt.xticks()
plt.yticks()
plt.show()
```

```python
k = int(input('Enter the number of clusters : '))
centroids = (data.sample(n=k))
plt.scatter(data['ApplicantIncome'],data['LoanAmount'],c = 'red')
plt.scatter(centroids['ApplicantIncome'],centroids['LoanAmount'], c = 'black')
plt.xlabel('Annual Income')
plt.ylabel('Loan amount (in thousands)')
plt.show()
```

```python
def distance(x1,y1,x2,y2):
    dist = ((x2 - x1)**2) + ((y2-y1)**2)
    dist = np.sqrt(dist)
    return dist
```

```python
x = data.values.tolist()
print(x)
print(len(x))
```

# Task 2 & 3: Partial table showing the data points, distance from each cluster andthe cluster assigned, Means Calculated after each iteration and A table showing the new cluster centers computed 

```python
c = centroids.values.tolist()
counter = 0
while True:
    counter = counter + 1
    print('Iteration number : {}'.format(counter))
    rt = []
    new_clusters = [[] for i in range(k)]
    bs = ['Cluster 1', 'Cluster 2', 'Cluster 3']
    togu = [[] for i in range(k)]
    rg = [[] for i in range(k)]
    c1 = 0
    for i in x:
        temp = []
        x1 = int(i[0])
        y1 = int(i[1])
        for j in c:
            x2 = int(j[0])
            y2 = int(j[1])
            t = distance(x1,y1,x2,y2)
            temp.append(t)
        m = min(temp)
        index_value = temp.index(m)
        temp.append(bs[index_value])
        rt.append(temp)
        rg[index_value].append(i)
        togu[index_value].append(m)

    df = DataFrame (rt,columns=['Cluster 1','Cluster 2','Cluster 3','Final Cluster'])
    print(df)
    print()

    for i in range(3):
        rg_count = 0
        sum1 = 0
        sum2 = 0
        for j in rg[i]:
            rg_count = rg_count + 1
            sum1 = j[0] + sum1
            sum2 = j[1] + sum2

        sum1 = sum1/rg_count
        sum2 = sum2/rg_count
        #print(sum1)
        #print(sum2)
        new_clusters[i].append(sum1)
        new_clusters[i].append(sum2)

    #print('Cluster centers : {} \n'.format(c))
    #print('Mean : {} \n'.format(new_clusters))
    
    print()
    print('Tabulated new cluster centers')
    print()
    df2 = DataFrame (new_clusters,columns=['ApplicantIncome','LoanAmount'])
    print(df2)
    print()
    print()
    
    if (c == new_clusters):
        break
    else:
        c = new_clusters
        togu.clear()
        rt.clear()
        rg.clear()
```

```python
df1 = DataFrame (new_clusters,columns=['ApplicantIncome','LoanAmount'])
print(df1)
```

# Task 4: Scatter plot showing final clusters with the center points

```python
for i in range(k):
    for j in rg[i]:
        if i == 0:
            plt.scatter(j[0],j[1],c = 'red')
        elif i == 1:
            plt.scatter(j[0],j[1],c = 'yellow')
        elif i == 2:
            plt.scatter(j[0],j[1],c = 'green')
plt.scatter(df1['ApplicantIncome'],df1['LoanAmount'], c = 'black')

plt.show()
```
