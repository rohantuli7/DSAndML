```python
import numpy as np
import pandas as pd
```

Learning rule used:

Updation of weights & biases:

if predicted y not equal to target

w(new) = w(old) + alpha*(input x)*(target)

b(new) = b(old) + alpha*(target)

if predicted y is equal to target

w(new) = w(old)

b(new) = b(old)

Activation function used:

TLU: thresholded logic unit

It uses the following:

f(y) = 1 (for y > 0)

f(y) = 0 (for y = 0)

f(y) = -1 (for y < 0)

# AND Gate

```python
#Input values for AND gate

x1 = np.array([1, 1, -1, -1])
x2 = np.array([1, -1, 1, -1])
y = np.array([1, -1, -1, -1])
w1 = 0
w2 = 0
b = 0
a = 1
t = 0
cols = ['X1', 'X2', 'bias', 'y', 'yin', 'Y', 'w1', 'w2', 'W1', 'W2']
df = pd.DataFrame(columns = cols)
```

```python
#Function for AND GATE
#This function uses perceptron learning rule to update weights and biases to obtain the correct output
#An activation function is also used

l = {}
for i in range(0, 4):
    l['X1'] = x1[i]
    l['X2'] = x2[i]
    l['y'] = y[i]
    l['bias'] = 1
    yin = b + (w1*x1[i]) + (w2*x2[i])
    l['yin'] = yin
    
    #Activation function
    if yin > 0:
        Y = 1
        l['Y'] = 1
    elif yin == 0:
        Y = 0
        l['Y'] = 0
    else:
        Y = -1
        l['Y'] = -1

    #Updation of weights wrt perceptron learning
    if Y == y[i]:
        l['w1'] = 0
        l['w2'] = 0
        l['W1'] = w1
        l['W2'] = w2
    else:
        w1 = w1 + (a*y[i]*x1[i])
        l['W1'] = w1
        b = b + (a*y[i])
        w2 = w2 + (a*y[i]*x2[i])
        l['W2'] = w2
        l['w1'] = a*y[i]*x1[i]
        l['w2'] = a*y[i]*x2[i]
    df = df.append(l, ignore_index = True)
```

## AND GATE OUTPUT

```python
display(df)
```
